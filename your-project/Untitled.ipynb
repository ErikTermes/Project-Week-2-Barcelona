### Rent price evolution per neighbourhood. 2019 - 2020 comparison

```python
import numpy as np
import pandas as pd
```

```python
rent_19 = pd.read_csv('../datasets/2019_lloguer_preu_trim.csv')
rent_20 = pd.read_csv('../datasets/2020_lloguer_preu_trim.csv')
```

```python
rent_19
```

```python
col_names_19 = ["Year", "Quarter", "District_Code", "District_Name", "Neighbor_Code", "Neighbor_Name", "Average_Rent", "Price"]
```

```python
rent_19.columns = col_names_19
```

```python
rent_19.drop(["District_Code","Neighbor_Code"], axis=1, inplace = True)
```

```python
rent_19["Average_Rent"].value_counts()
```

```python
e_mes_19 = rent_19[rent_19["Average_Rent"] == "Lloguer mitjà mensual (Euros/mes)"]
```

```python
e_mes_19
```

```python
wo_Q1_19 = e_mes_19.drop(e_mes_19[e_mes_19.Quarter == 1].index)
wo_Q3_19 = wo_Q1_19.drop(wo_Q1_19[wo_Q1_19.Quarter == 3].index)
Q2_price_19 = wo_Q3_19.drop(wo_Q3_19[wo_Q3_19.Quarter == 4].index)
Q2_price_19
```

```python
Q2_price_19["District_Name"].value_counts()
```

```python
Q2_price_19["Price"].mean()
```

```python
price_by_hood_19 = Q2_price_19.groupby(["District_Name","Neighbor_Name"])["Price"].mean()
price_by_hood_19
```

```python
price_by_district_19 = Q2_price_19.groupby(["District_Name"])["Price"].mean()
price_by_district_19
```

```python
Q2_District_Mean19 = Q2_price_19.groupby(["District_Name"]).mean().sort_values("Price", ascending = False)
Q2_District_Mean19
```

```python
Q2_District_Mean19
```

```python
Q2_District_Mean19.drop(["Quarter","Year"], axis=1, inplace = True)
```

```python
Q2_District_Mean19
```

```python
min_price_by_district_19 = Q2_price_19.groupby(["District_Name"])["Price"].min()
min_price_by_district_19
```

```python
max_price_by_district_19 = Q2_price_19.groupby(["District_Name"])["Price"].max()
max_price_by_district_19
```

```python
Q2_price_19["Price"].mean()
```

```python
std_price_by_district_19 = Q2_price_19.groupby(["District_Name"])["Price"].std()
std_price_by_district_19
```
